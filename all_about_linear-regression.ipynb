# Goal of Linear Regression

To find out the best fitting line between all datapoints is the main motto of Linear Regression. This best fitting line minimises the difference between the actual and predictive value of y. Linear Regression done this task by appling least squares method which minimizes the sum of the squared differences between the observed and predicted values.

So, in essence, linear regression involves finding the parameters (slope m and intercept c) of a linear equation that best fits the given data points, thus establishing a relationship between the variables that can be represented by a straight line.

```python
# import the necessary library 
import pandas as pd   
import numpy as np
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/height-weight/Height-Weight Data.csv")
df.head()
```

# Visualize the data

```python
plt.scatter(df.Height,df.Weight, color='green', marker='*')
plt.ylabel('Weight')
plt.xlabel('Height')
plt.title('Height-Weight Data')
```

# Spliting the dataset

```python
x =df[['Height']] 
y= df[['Weight']]
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest = tts(x,y, test_size= .20, random_state= 42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
lr.score(xtrain,ytrain)
```

```python
lr.score(xtest,ytest)
```

# Set an individual value applying the equation to get the value of y 

```python
m = lr.coef_
c = lr.intercept_
```

Suppose here the variable predict_value represents the value of y

```python
Predict_value= (m*65.118504) + c
Predict_value
```

```python
df.iloc[160:171]
```

```python
df['predicted']= lr.predict(df[['Height']])
df.head()
```

# Draw the best fit line for the test data

```python
plt.scatter(df[['Height']], df[['Weight']], color='green', marker='*') 
plt.plot(xtest, lr.predict(xtest), color='red')
plt.ylabel('Weight')
plt.xlabel('Height')
plt.title('Height-Weight Data')
```

# Error calculation

By calculating the error we can findout the model effectiveness. Error is particularly calculated to find out the difference between the predicted value of y and the actual value of y.

```python
from sklearn.metrics import mean_squared_error as MSE    
from sklearn.metrics import mean_absolute_error as MAE
```

```python
test_mse= MSE(ytest, lr.predict(xtest))
test_mse
```

```python
test_mae= MAE(ytest, lr.predict(xtest))
test_mae
```
